# Lets deep dive into our analysis to gain more insights

# Univariate Analysis

```python
fig, axis = plt.subplots(nrows = 2 , ncols = 2 , figsize = (15 , 7))
sns.countplot(data = df , x = df["Ship Mode"] , ax = axis[0 , 0])  
sns.countplot(data = df , x = df["Segment"] , ax = axis[0 , 1])  
sns.countplot(data = df , x = df["Region"] , ax = axis[1, 0])  
sns.countplot(data = df , x = df["Category"] , ax = axis[1, 1])
plt.tight_layout()
plt.show()
```

# OBSERVATIONS

1. Most of the orders have ship mode as Standard Class and least for the Same Day.
2. Majority of the orders are Consumer segment while least for Home Office.
3. Highest number of orders are from West and East Region while least from South.
4. Majorily the category of order are of Office supplies as compared to other categories. 

```python
fig, axis = plt.subplots(nrows = 4 , ncols = 1 , figsize = (15 , 7))
sns.boxplot(data = df , x = df["Sales"] , ax = axis[0],color = "g")  
sns.boxplot(data = df , x = df["Quantity"] , ax = axis[1],color = "r")  
sns.boxplot(data = df , x = df["Discount"] , ax = axis[2],color = "b")  
sns.boxplot(data = df , x = df["Profit"] , ax = axis[3],color = "y")
plt.tight_layout()
plt.show()
```

# OBSERVATIONS

1. There are many outliers in Sales attribute and all of them are in positive side i.e higher side.
2. Majority of the Quantity lies between 3 to 9 orders with some outliers on higher side.
3. Discount majorily ranges from 0 to 50% with 60%,70% and 80% as an outlier.
4. Profit has highest number of outliers and seems to incur losses too.

# Bi-Variate Analysis

```python
df.head()
```

```python
pd.DataFrame(df.groupby('Ship Mode').sum()[['Sales','Profit']]).plot(kind='bar',figsize=(6,3))
pd.DataFrame(df.groupby('Region').sum()[['Sales','Profit']]).plot(kind='bar',figsize=(6,3))
pd.DataFrame(df.groupby('Segment').sum()[['Sales','Profit']]).plot(kind='bar',figsize=(6,3))
pd.DataFrame(df.groupby('Category').sum()[['Sales','Profit']]).plot(kind='bar',figsize=(6,3))
plt.tight_layout()
plt.show()
```

# OBSERVATIONS

1. The profit is far less as compared to the sales that have been made through Standard Class ship mode that infers that some good amount of losses have been incurred.
2. The sales are lowest from south region.
3. Profits in central region are not upto the mark as compared to other regions.
4. Sales and Profit both are highest for consumer segment and lowest for Home office.
5. Profit made in furniture category are extremely low as compared to other categories.

```python
pd.DataFrame(df.groupby('Sub-Category').sum()[['Profit']]).plot(kind='bar',color='r',figsize=(6,4))
```

```python
pd.DataFrame(df.groupby('Sub-Category').sum()[['Sales']]).plot(kind='bar',color = 'g',figsize=(6,4))
```
